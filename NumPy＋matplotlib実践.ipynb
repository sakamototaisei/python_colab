# **NumPyとmatplotlibの概要**

## **NumPyとmatplotlibの実演**

**データの生成**

NumPyを使ってデータを生成、加工する

-πからπまでの数値データを生成した上で、簡単な演算を行う

```python
import numpy as np

x = np.linspace(-np.pi, np.pi)
y_1 = 0.5 * x
y_2 = np.sin(x)

print('x')
print(x)
print()
print('y=0.5x')
print(y_1)
print()
print('y=sin x')
print(y_2)
```

**グラフの描画**

matplotlibを使ってグラフを描画する

軸のラベルグラフのタイトル、凡例などを表示し、線のスタイルを変更してリッチなグラフにする

```python
import matplotlib.pyplot as plt

# 軸のラベル
plt.xlabel('x value')
plt.ylabel('y value')

# グラフのタイトル
plt.title('My Graph')

# プロット凡例と線のスタイルを指定
plt.plot(x, y_1, label='0.5x')
plt.plot(x, y_2, label='sin x', linestyle='dashed')
plt.legend()

plt.show()
```
